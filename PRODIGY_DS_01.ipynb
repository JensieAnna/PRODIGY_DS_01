#### Plotting the bar chart for Workclass Distribution

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='workclass', data=df, palette='Set2')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Workclass Distribution')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Plotting the histogram for Age Distribution

```python
plt.figure(figsize=(10, 6))
sns.distplot(df['age'], bins=20,color='skyblue', hist_kws={'edgecolor': 'black', 'linewidth': 1.2})
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age Distribution')
plt.tight_layout()
plt.show()
```

#### Plotting the bar chart for Education Level Distribution

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='education', data=df, palette='Set3')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.title('Education Level Distribution')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Plotting the bar chart for Occupation Distribution

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='occupation', data=df, palette='viridis')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Occupation Distribution')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Plotting the bar chart for Hours Per Week Distribution by Income Level

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='income', y='hours-per-week', data=df, palette='coolwarm')
plt.xlabel('Income Level')
plt.ylabel('Hours Per Week')
plt.title('Hours Per Week Distribution by Income Level')
plt.tight_layout()
plt.show()
```

#### Plotting the bar chart for Education level distribution by Gender

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='education', hue='gender', data=df, palette='pastel')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.title('Education Level Distribution by Gender')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
